````
# 나의 과제그룹 : Task 1 (3-class Classification): Predict the diagnosis group of subjects
# 예측함수 작성 : def read_csv_and_predict(csv_path)

from google.colab import files
import joblib

import pandas as pd
import numpy as np
np.random.seed(20220513)

# 선정된 모델library import
from sklearn.ensemble import RandomForestClassifier

from sklearn.preprocessing import MinMaxScaler #sklearn에서 제공하는 Feature Scaling Library

model_files = files.upload() # You can upload multiple files at once. If there are any problems uploading, try uploading to Colab directly using the file explorer in the sidebar.

# 나의 과제그룹 : Task 1 (3-class Classification): Predict the diagnosis group of subjects
clf = joblib.load("clf_2020511021.pkl") # 사전에 피팅한 모델 로딩
#reg1 = joblib.load("reg1.pkl")
#reg2 = joblib.load("reg2.pkl")
#reg3 = joblib.load("reg3.pkl")

def read_csv_and_predict(csv_path):
    import pandas as pd
    train_df = pd.read_csv(csv_path)
    X_all = train_df[[c for c in train_df.columns if c not in ['DX_bl', 'ADAS11', 'ADAS13', 'MMSE']]]

    # 데이터 확인
    print( "X_all.shape=",X_all.shape )
    print( X_all.value_counts() ) # 값 별로 건수 표시
    #print( X_all ) # 전체 데이터 확인
    #print( X_all.describe() ) #데이터 count, mean, std, min, 25%,50%,75%, max 값 표시

    # 데이터 전처리 ① : NaN --> 0 --> mean값으로 채움

    X0 = X_all.fillna(0).copy() # 0을 채운 후 복사하여 사용, 아래에서 mena값을 위해
    print("'X_all' contain NaN values ?",np.any(np.isnan(X_all) ) ) # NaN값 있는지 확인
    print("'X0' contain NaN values ?",np.any(np.isnan(X0) ) ) # NaN값 있는지 확인
    Xm = X_all.fillna( X0.mean() ).copy() # NaN값을 평균값으로 채움, 위에서 0을 채운값으로 평균값 계산
    print("'X_all' contain NaN values ?",np.any(np.isnan(X_all) ) ) # NaN값 있는지 확인
    print("'Xm' contain NaN values ?",np.any(np.isnan(Xm) ) ) # NaN값 있는지 확인

    # 데이터 전처리 ② : scaling
    # MinMax scale features into 0~1 (e.g., use MinMaxScaler)
    scaler1 = MinMaxScaler() # 객체 생성
    scaler1.fit(Xm) # 피팅 : 스케일링에 사용할 최소,최대 값 계산하여 저장
    Xm_scaled = scaler1.transform(Xm) # 데이터 변환, 결과는 배열로 반환됨
    X = np.round(Xm_scaled,4) # 결과값들을 소수점 4째자리까지 반올림 (불필요값 제거)
    print("My X data range :", np.min(X) , "~", np.max(X) )

    predictions = {}
    predictions["DX_bl"] = clf.predict_proba(X) # 예측 확률값
    arr_cnt = len(predictions["DX_bl"])
    predictions["ADAS11"] = np.zeros(arr_cnt) #reg1.predict(X)
    predictions["ADAS13"] = np.zeros(arr_cnt) #reg2.predict(X)
    predictions["MMSE"]   = np.zeros(arr_cnt) #reg3.predict(X)

    print(predictions)

    return predictions
````
